## Conclusiones

Objetivo del proyecto: Identificar el perfil de los usuarios que generan mayor tiempo de visualización en la plataforma, para aportar información que permita diseñar estrategias de retención e incremento de ingresos.

Pregunta de investigación: ¿Qué características tienen los usuarios que consumen más tiempo de contenido en la plataforma?


1. Distribución general del consumo

- Evidencia: El histograma de minutos de visualización mensual mostró una distribución asimétrica a la derecha, con la mayoría de los usuarios concentrados entre 500 y 1200 minutos (media = 793.9, mediana = 758.5), y un grupo minor ubicado entre 2500 y 4300 minutos.
- Interpretación: La separación entre ambos grupos, en lugar de una cola continua, sugiere que el alto consumo no es simplemente "un poco más de uso", sino un patrón de comportamiento distinto.
- Conclusión: Existe un segmento diferenciado de usuarios de alto consumo que debe ser tratado como un perfil diferente antes que como un extremo estadístico de la distribución general.

2. Relación entre edad y consumo

- Evidencia: El gráfico de dispersión edad vs. minutos mostró que el consumo elevado se concentra principalmente entre los 15 y 60 años, y disminuye notablemente después de esa edad.
- Interpretación: La edad actúa como un factor limitante más que determinante: fuera de ese rango el alto consumo es infrecuente, pero dentro de él no permite diferenciar con precisión quién pertenece a cada grupo.
- Conclusión: La edad es una variable relevante pero insuficiente por sí sola para explicar el perfil de alto consumo.

3. Relación entre plan de suscripción y consumo

- Evidencia: El promedio de minutos por plan mostró una relación creciente (Básico = 600, Estándar = 900, Premium = 1050 minutos), pese a que el plan Básico concentra la mayor cantidad de usuarios. El gráfico multivariado (edad, plan y minutos) evidenció además que los tres planes tienen usuarios presentes tanto en el grupo de consumo moderado como en el de alto consumo.
- Interpretación: El plan contratado está asociado al nivel de consumo, pero no lo determina de forma exclusiva: existen usuarios de planes económicos con un comportamiento de consumo igual al de usuarios Premium.
- Conclusión: El plan es un indicador parcial del perfil de alto consumo. El hallazgo más relevante para el objetivo de negocio es la existencia de usuarios Básico/Estándar con consumo elevado, que constituyen candidatos naturales para estrategias de marketing y ofrecerles un plan mejor.

4. Exploración de componentes principales (PCA)

- Evidencia: El análisis de PCA sobre las variables edad, minutos de visualización y tickets de soporte al cliente mostró que se requieren los tres componentes para explicar aproximadamente el 100% de la varianza, sin una reducción significativa de dimensionalidad. El CP1 contrapone consumo de contenido y tickets de soporte, CP2 está dominado por la edad, y CP3 combina las tres variables.
- Interpretación: La baja correlación entre las variables originales impide que el PCA concentre la varianza en pocos componentes, aunque permitió identificar una posible asociación inversa entre tiempo de consumo y cantidad de tickets de soporte.
- Conclusión: El PCA no resultó eficaz como técnica de reducción de dimensionalidad en este caso, pero aportó valor exploratorio al sugerir que el consumo elevado podría estar asociado a una menor cantidad de reclamos de soporte, una relación que podría profundizarse en análisis posteriores.

Conclusión general

En base a la evidencia recolectada, los usuarios que consumen más tiempo de contenido en la plataforma tienden a ubicarse en un rango etario de 15 a 60 años y muestran mayor probabilidad de pertenecer a planes de suscripción superiores, aunque el alto consumo también se manifiesta de forma significativa en usuarios de planes económicos. Esto indica que el comportamiento de consumo elevado responde a un patrón de uso que no está determinado exclusivamente por el plan contratado, lo que abre una oportunidad concreta de negocio: identificar y convertir en usuarios premiun a los usuarios de planes básicos que ya presentan un consumo equivalente al de usuarios premium.


# Limitaciones y mejoras futuras

Limitaciones

- El alcance de las conclusiones se encuentra condicionado por las variables disponibles en el dataset (edad, plan, minutos de visualización, tickets de soporte), que no permiten explicar completamente la variabilidad del comportamiento de consumo.
- El análisis fue de carácter exploratorio y descriptivo; las relaciones identificadas entre variables no fueron validadas mediante modelos predictivos.
- El PCA aplicado no logró reducir efectivamente la dimensionalidad del conjunto de datos, dado el bajo nivel de correlación entre las variables numéricas disponibles.

Mejoras futuras

- Una mejora futura podría consistir en incorporar variables adicionales (dispositivo de acceso, género de contenido preferido, antigüedad como usuario) que permitan enriquecer el perfil del usuario de alto consumo.
- Se propone como siguiente etapa el desarrollo de un modelo de clasificación supervisada que permita predecir la probabilidad de que un usuario pertenezca al segmento de alto consumo a partir de las variables identificadas como relevantes durante el análisis exploratorio.
- Sería recomendable explorar con mayor profundidad la relación entre tiempo de consumo y tickets de soporte identificada en el PCA, dado su potencial valor para estrategias de retención.
